```python
x=data[['FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB']]
y=data['CO2EMISSIONS']
#x.values
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_tt,y_tr,y_tt=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_tr.shape
```

```python
x_tt.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_tr,y_tr)
```

```python
y_pred=model.predict(x_tt)
```

```python
y_tt.head(1)
```

```python
y_pred[0]
```

```python
model.score(x,y)
```
